```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rows = []
with open("final.csv") as f:
  csvreader = csv.reader(f)
  for row in csvreader:
    rows.append(row)

stars_data_rows = rows[0:]
print(stars_data_rows[1])
stars_data = stars_data_rows[0:]
```

```python
headers = rows[0]
headers[0] = "index"
print(headers)
```

```python
df = pd.read_csv("final.csv")

star_name = df["Star_name"].to_list()
distance = df["Distance"].to_list()
mass = df["Mass"].to_list()
radius = df["Radius"].to_list()
gravity = df["Gravity"].to_list()
```

```python
plt.figure(figsize = (20, 5))
plt.title("Star vs Distance")
plt.xlabel("Star Name")
plt.ylabel("Distance")
plt.bar(star_name, distance)
```

```python
plt.figure(figsize = (20, 5))
plt.title("Star vs Mass")
plt.xlabel("Star Name")
plt.ylabel("Mass")
plt.bar(star_name, mass)
```

```python
plt.figure(figsize = (20, 5))
plt.title("Star vs Radius")
plt.xlabel("Star Name")
plt.ylabel("Radius")
plt.bar(star_name, radius)
```

```python
plt.figure(figsize = (20, 5))
plt.title("Star vs Gravity")
plt.xlabel("Star Name")
plt.ylabel("Gravity")
plt.bar(star_name, gravity)
```
